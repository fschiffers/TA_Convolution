# Let's try this on 2D images

```python
img_orig = misc.face()
print('Size of loaded img: ' + str(img_orig.shape))
```

```python
plt.imshow(img_orig)
plt.show()
```

Do some preprocessing

```python
# Convert to gray scale
img = skimage.color.rgb2gray(img_orig)
```

```python
crop_cen = [450,600]
N_img = 512
crop_rad = (int)(N_img / 2)
```

```python
img = img[crop_cen[0]-crop_rad:crop_cen[0] + crop_rad, crop_cen[1]-crop_rad:crop_cen[1] + crop_rad]
print("New image shape: " +str(img.shape))
# Rescale if needed
#img = skimage.transform.rescale(img,[0.5,0.5])
```

```python
plt.imshow(img,cmap='gray')
plt.show()
```

# create disk-like filter footprint with given radius

```python
x = np.linspace(-N_img/2,N_img/2,N_img)
[X,Y] = np.meshgrid(x,x)
R = np.sqrt(X**2 + Y**2)
```

```python
plt.imshow(R)
plt.colorbar()
plt.title("Radial distance from the center")
plt.show()
```

```python
# Defines the size of the gaussian PSF in Pixel
sigma = 15
print(sigma)
psf_gauss = np.exp(-( X**2 + Y**2) / (2*sigma**2))
#(sigma * np.sqrt(2*np.pi))
```

```python
plt.imshow(X**2 + Y**2)
plt.imshow(psf_gauss)
plt.show()
```

```python
radius = 11.5
disk = np.zeros(R.shape)
R_disk = 10
disk[R<R_disk] = 1
```

```python
# Alternative
disk_PSF = skimage.morphology.disk(R_disk)
plt.imshow(disk_PSF)
plt.show()
```

```python
# Show disk
plt.imshow(disk)
plt.show()
```

```python
# Let's calculate the convolution according to formula above
im_filtered = scipy.ndimage.convolve(img,disk_PSF,mode='reflect')
```

```python
plt.figure(figsize=(14,10))
plt.subplot(121)
plt.imshow(img, cmap = 'gray')
plt.title('Original')
plt.subplot(122)
plt.imshow(im_filtered, cmap = 'gray')
plt.title('Filtered')
plt.show()
```

# Convolution theorem ( Very Important ! )

<h1><center>$ F(f * g) = F(f) \cdot F(g) $</center></h1>

Why is this cool ?!
<br>
Because Fourier transforms are faster (!) than convolution (especially for large kernels)

# Let's try to implement the convolution theorem

## Pad filter to image size (The PSF/OTF need to be the size of the image)

```python
pad_size = (np.asarray(img.shape) - np.asarray(disk.shape) )/ 2
pad_size = pad_size.astype(int)
print(pad_size)
disk_pad = np.lib.pad(disk, pad_size, 'constant',constant_values=0)
print(disk_pad.shape)
```

# Check that padding has worked

```python
plt.imshow(disk_pad)
plt.show()
```

# Calculate frequency response

The optical transfer function (OTF) is the Fourier Transform of the Point Spreaf Function (PSF)

```python
otf = np.fft.fft2(disk_pad)
```

```python
# Display OTF
plt.imshow(np.abs(otf))
plt.show()
```

# Looks weird ?! We need to FFTSHIFT (FFT2 has this weird property ...)

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(np.log(np.abs(np.fft.fftshift(otf))+1))
plt.title('Optical Transfer Function')
plt.subplot(122)
y = (np.abs(np.fft.fftshift(otf)[256,:]))
y = y/max(y)
plt.plot(x,y)
plt.xlabel('Relative frequency')
plt.ylabel('Normalized MTF')
plt.title('Optical Transfer Function')
plt.show()
```

## Then we calculate the "spectrum" of the object

```python
im_fft2 = np.fft.fft2(img)
```

```python
# Let's visualize the spectrum
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(img)
plt.title('Image')
plt.subplot(122)
plt.imshow(np.log(np.abs(np.fft.fftshift(im_fft2))+1))
plt.title('Object spectrum')
plt.show()
```

```python
# According to the convolution theorem we multiply both images elementwise
```

```python
im_filtered_fft2 = im_fft2 * otf
```

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.imshow(np.log(np.abs(np.fft.fftshift(im_fft2))+1))
plt.title('Spectrum Object')
plt.subplot(132)
plt.imshow(np.log(np.abs(np.fft.fftshift(otf))+1))
plt.title('OTF')
plt.subplot(133)
plt.imshow(np.log(np.abs(np.fft.fftshift(im_filtered_fft2))+1))
plt.title('Filtered Spectrum')

plt.show()
```

# Now we need perform inverse FFT to obtain the filtered image

```python
im_filtered = np.real(np.fft.fftshift(np.fft.ifft2(im_filtered_fft2)))
# Normalize image
im_filtered = im_filtered / np.max(im_filtered)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(img)
plt.title('Original')

plt.subplot(122)
plt.imshow(im_filtered)
plt.title('Filtered through Convolution theorem')
plt.show()
```

# Can we restore the image ? 
## Yes, that's what we call: Deconvolution

<h1><center> $ y = h * x = A \cdot x$ </center></h1>

with h being the convolution kernel, e.g. a gaussian

A convolution can always (!) be written as a matrix multiplication with block-circulant matrix $A$

<h1><center> Hence we only have to invert matrix A<\center></h1>

<h1><center> $ x = A^{-1} \cdot y $ </center></h1>

How to solve for x ?

One could try to calculate $A^{-1}$ directly, but you will quickly realize that this matrix is infeasible to invert for large dimensions.

Instead, one could e.g. think of the convolution theorem. 
<h1><center><br>
$y = h * x$
<br>
$Y = H \cdot X$
<br>
$X = Y ./ H$
</center></h1>

<center><h1>Deconvolution boils down to pointwise dvision</h1></center>

```python
# Calculate fourier spectrum
im_filtered_fft2 = np.fft.fft2(im_filtered)
```

```python
# Set frequencies where OTF is to 0 !
im_deconv_fft2 = np.divide(im_filtered_fft2, otf, out=np.zeros_like(otf), where=otf!=0)
```

```python
im_deconv = np.real(np.fft.fftshift(np.fft.ifft2(im_deconv_fft2)))
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(im_filtered)
plt.title('Filtered')

plt.subplot(122)
plt.title('Deconvolved')
plt.imshow(im_deconv)

plt.show()
```

<h1><center>What happens if we add noise ? </center></h1>

```python
noise_sigma = 0.0001
noise = noise_sigma*np.random.randn(im_filtered.shape[0],im_filtered.shape[1])
im_filtered_noisy = im_filtered + noise
```

```python
def colorbar(mappable):
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    import matplotlib.pyplot as plt
    last_axes = plt.gca()
    ax = mappable.axes
    fig = ax.figure
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    cbar = fig.colorbar(mappable, cax=cax)
    plt.sca(last_axes)
    return cbar
```

```python
fig,(ax1, ax2, ax3) = plt.subplots(ncols=3,figsize=(15,15))

img1 = ax1.imshow(noise,cmap='gray')
ax1.set_title('Noise')
colorbar(img1)

img2 = ax2.imshow(im_filtered_noisy,cmap='gray')
ax2.set_title('Noisy image')
colorbar(img2)


img3 = ax3.imshow(im_filtered,cmap='gray')
ax3.set_title('Noisy image')
colorbar(img3)

plt.tight_layout(h_pad=1)
plt.show()
```

<h1><center> ! ! ! Let's do the deconvolution again ! ! ! 
<br>
 Nothing has changed,
 <br>
 ...  or has it ?  </h1></center> 

```python
im_filtered_fft2 = np.fft.fft2(im_filtered_noisy)
im_deconv_fft2 = np.divide(im_filtered_fft2, otf, out=np.zeros_like(otf), where=otf!=0)
im_deconv = np.real(np.fft.fftshift(np.fft.ifft2(im_deconv_fft2)))
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(im_filtered)
plt.title('Filtered')

plt.subplot(122)
plt.title('Deconvolved')
plt.imshow(im_deconv)

plt.show()
```

<h1><center> ? ? ? WHAT HAPPENED ? ? ? </h1></center>

<h1><center> Divison by values close to 0. We need to regularize </h1></center>

<h1><center> $ Y = \frac{X}{H + \epsilon}$ </h1></center>

```python
im_filtered_fft2 = np.fft.fft2(im_filtered)
```

```python
eps_regularizer = 0.25
im_filtered_fft2 = np.fft.fft2(im_filtered_noisy)
im_deconv_fft2 = im_filtered_fft2/(otf + eps_regularizer)
im_deconv = np.real(np.fft.fftshift(np.fft.ifft2(im_deconv_fft2)))
```

```python
plt.imshow(im_deconv)
plt.title('Thikonov Regularized Image')
plt.show()
```

# Better way, write problem as inverse problem:

<center><h1>  $\hat{x} = \underset{x}{\operatorname{arg min}} L(x) $
<br>
$ L(x) = \frac{1}{2} || y - A\cdot x||_2^2 + \phi(x)$ 
<br>
 </h1>
, where $ \phi(x) $ is a (differentiable) regularization term,
<br> e.g. $ \phi(x) = \sum\limits_{k=1}^{K}||x_k||^2 $
<br>
This can be solved either iteratively (gradient descent) or in some cases even directly
</center>
